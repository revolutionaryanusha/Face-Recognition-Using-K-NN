```python
import pandas as pds
import numpy as npy
from collections import Counter

train_X , train_y , test_X , test_y  = [],[],[],[]

df = pds.read_csv('dataset.csv', header=None)
df_norm = npy.linalg.norm(df, axis=1, keepdims=True)
df_normalized = df / df_norm

i = 0
const=170
const1=150
flag=False
while i < 10:
    idx = np.random.permutation(const)
    train_idx = idx[:const1]
    test_idx = idx[const1:]
    if flag== False:
        train_X.extend(df_normalized.iloc[i * const + train_idx].values)
        train_y.extend([i] * len(train_idx))
        test_X.extend(df_normalized.iloc[i * const + test_idx].values)
        test_y.extend([i] * len(test_idx))
    i += 1

k_vals = [2, 5, 7, 11]
distance_methods = ['euclidean', 'cosine']
for k in k_vals:
    for method in distance_methods:
        predictions = []
        for i in range(len(test_X)):
            dists = []
            for j in range(len(train_X)):
                if method == 'euclidean':
                    d = npy.linalg.norm(test_X[i] - train_X[j])
                elif method == 'cosine':
                    d = 1 - npy.dot(test_X[i], train_X[j]) / (npy.linalg.norm(test_X[i]) * npy.linalg.norm(train_X[j]))
                dists.append((d, train_y[j]))
            dists.sort()
            nearest_neighbors = [y for _, y in dists[:k]]
            prediction = Counter(nearest_neighbors).most_common(1)[0][0]
            predictions.append(prediction)
        accuracy = sum(npy.array(predictions) == npy.array(test_y)) / len(test_y)
        print(f'k={k}, method={method}, accuracy={accuracy:.2f}')
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

svm_clf = SVC()
svm_clf.fit(train_X, train_y)
svm_predictions = svm_clf.predict(test_X)
svm_accuracy = accuracy_score(test_y, svm_predictions)
print(f'SVM accuracy={svm_accuracy:.2f}')


gnb_clf = GaussianNB()
gnb_clf.fit(train_X, train_y)
gnb_predictions = gnb_clf.predict(test_X)
gnb_accuracy = accuracy_score(test_y, gnb_predictions)
print(f'GaussianNB accuracy={gnb_accuracy:.2f}')
```
